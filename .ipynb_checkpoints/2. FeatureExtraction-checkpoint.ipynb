```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
try:
    import dill as pickle
except:
    !pip install dill
    import dill as pickle
from tqdm import tqdm
from src import DataTransforms
%matplotlib inline
```

```python
%%time
azdias = pd.read_csv('../data/Udacity_AZDIAS_052018.csv', sep=';', low_memory = False, index_col = 'LNR')
```

```python
with open('artifacts/transforms.pth','rb') as f:
    transforms = pickle.load(f)
```

```python
azdias = DataTransforms.TransformData(azdias, transforms)
```

```python
from sklearn.manifold import TSNE
```

```python
%%time
#create TNSE visualization using random subsets of the data, can't use the whole dataset because it's huge
nrow = 4
ncol = 3
fig, axis = plt.subplots(nrow, ncol)
for k in tqdm(range(nrow*ncol)):
    X = TSNE(n_components=2).fit_transform(azdias.sample(1000).values)
    
    i = k//ncol
    j = k %ncol
    axis[i][j].scatter(X[:,0], X[:, 1])
fig.set_figheight(15)
fig.set_figwidth(20)
fig.savefig('tSNE.png')
plt.show(fig)
```

```python
data = azdias.values
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
%%time
#fit PCA
pca.fit(data)
```

```python
explained_variance = 0
k = 0
while explained_variance < 0.98 and k < len(pca.explained_variance_ratio_):
    explained_variance += pca.explained_variance_ratio_[k]
    k += 1

#k is the least number of features that expains 0.98 of the variance of the data
print(f'k = {k}')
print(pca.explained_variance_ratio_[k-1])
```

```python
#train a new PCA with k
pca = PCA(k)
```

```python
%%time
pca.fit(data)
```

```python
with open('artifacts/pca.pkl' ,'wb') as f:
    pickle.dump(pca, f)
```
