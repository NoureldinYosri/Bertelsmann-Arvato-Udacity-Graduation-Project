```python
from src import DataTransforms


#transforms[col] is a list of transformation that has to be applied to the col in order
transforms = {k:[] for k in azdias.columns}
```

```python
to_drop = []
#explore categorical data and drop columns with small std or high percent of Nan
for col in azdias.columns:
    nan_percent = azdias[col].isna().sum() / azdias.shape[0] * 100
    if nan_percent > 40:
        #if +40% of the column in NaN then ignore this column
        transforms[col].append(DataTransforms.Drop())
        to_drop.append(col)
        continue
    if azdias[col].dtype == "int64" or azdias[col].dtype == "float64":
        if azdias[col].std() < 0.5:
            # if the std < 0.5 (e.g. most enteries have approximatly the same value) then ignore the column 
            transforms[col].append(DataTransforms.Drop())
            to_drop.append(col)
        continue
            
    print(col, azdias[col].dtype, flush = True)
    print(azdias[col].describe())
    print(f'NaN percent\t{nan_percent : 0.2f}%')
    print(azdias[col].unique())
    print(flush = True)
```

```python
#drop creation time (EINGEFUEGT_AM)
to_drop.append('EINGEFUEGT_AM')
transforms['EINGEFUEGT_AM'].append(DataTransforms.Drop())
```

```python
#in OST_WEST_KZ fill na with W as it's the most frequent with a hugh margin
azdias['OST_WEST_KZ'] = azdias['OST_WEST_KZ'].fillna('W')
transforms['OST_WEST_KZ'].append(DataTransforms.Replace(None, 'W'))
```

```python
#replace 'W' with 1 and 0 otherwise
azdias['OST_WEST_KZ'] = (azdias['OST_WEST_KZ'] == 'W').astype('int32')
azdias['OST_WEST_KZ'].head()
transforms['OST_WEST_KZ'].append(DataTransforms.CustomTransform(lambda s: int(s == 'W')))
```

```python
#D19_LETZTER_KAUF_BRANCHE fill na with unknown (D19_UNBEKANNT)
#since D19_LETZTER_KAUF_BRANCHE is has no valid total ordering I'll one-hot encode it


data = azdias['D19_LETZTER_KAUF_BRANCHE'].fillna('D19_UNBEKANNT')
to_drop.append('D19_LETZTER_KAUF_BRANCHE')
one_hot = pd.get_dummies(data)
for col in one_hot.columns:
    azdias[col] = one_hot[col]
    
transforms['D19_LETZTER_KAUF_BRANCHE'].append(DataTransforms.Replace(None, 'D19_UNBEKANNT'))
transforms['D19_LETZTER_KAUF_BRANCHE'].append(DataTransforms.OneHotTransform(data))

```

```python
#CAMEO_DEUG_2015 is the first char (number) of CAMEO_DEU_2015
#where 1 is the richest, 10 is the poorest and X is unknown
azdias["CAMEO_DEUG_2015"] = azdias["CAMEO_DEUG_2015"].fillna('X')
azdias['CAMEO_DEUG_2015'] = azdias['CAMEO_DEUG_2015'].replace('X', '50') # special value
azdias['CAMEO_DEUG_2015'] = azdias['CAMEO_DEUG_2015'].astype('int64')

transforms['CAMEO_DEUG_2015'].append(DataTransforms.Replace(None, 'X'))
transforms['CAMEO_DEUG_2015'].append(DataTransforms.Replace('X', '50'))
```

```python
#CAMEO_DEU_2015 will have only the char
azdias['CAMEO_DEU_2015'] = azdias['CAMEO_DEU_2015'].fillna('XX')
azdias['CAMEO_DEU_2015'] = azdias['CAMEO_DEU_2015'].apply(lambda s: ord(s[1]) - ord('A'))

transforms['CAMEO_DEU_2015'].append(DataTransforms.Replace(None, 'XX'))
transforms['CAMEO_DEU_2015'].append(DataTransforms.CustomTransform(lambda s: ord(s[1]) - ord('A')))
```

```python
#CAMEO_INTL_2015 also classfiy from reachiest (1) to poorest (55) where XX is unkown
azdias['CAMEO_INTL_2015'] = azdias['CAMEO_INTL_2015'].fillna('XX')
azdias['CAMEO_INTL_2015'] = azdias['CAMEO_INTL_2015'].replace('XX', '100')
azdias['CAMEO_INTL_2015'] = azdias['CAMEO_INTL_2015'].astype('int64')

transforms['CAMEO_INTL_2015'].append(DataTransforms.Replace(None, 'XX'))
transforms['CAMEO_INTL_2015'].append(DataTransforms.Replace('XX', '100'))
```

```python
azdias = azdias.drop(columns = to_drop)
```

```python
#get a list of columns with hidden values
numerical_types = ['int64', 'float64']
columns_with_na = []
for col in azdias.columns:
    if azdias[col].dtype not in numerical_types: continue
    if not azdias[col].isna().sum(): continue
    columns_with_na.append(col)
print(len(columns_with_na))
print(columns_with_na)
```

```python
#for each col record its mean
for col in columns_with_na:
    print(col, azdias[col].dtype, flush = True)
    print(azdias[col].describe())
    print(f'NaN percent\t{azdias[col].isna().sum() / azdias.shape[0] * 100 : 0.2f}%')
    print(azdias[col].unique())
    mu = azdias[col].mean()
    transforms[col].append(DataTransforms.Replace(None, mu))
    print('replace na with',mu)
    azdias[col] = azdias[col].fillna(mu)
    print(flush = True)
```

```python
for col in azdias.columns:
    if azdias[col].isna().any():
        print(col)
        print(azdias[col].describe())
        print(azdias[col].isna().sum() / azdias.shape[0] * 100)
```

```python
try:
    from tqdm import tqdm
except:
    !pip install tqdm
    from tqdm import tqdm
from sklearn import preprocessing


#normalize input to the range [0,1]
for col in azdias.columns:
    min_max_scaler = preprocessing.MinMaxScaler()
    min_max_scaler.fit(azdias[col].values.reshape(-1,1))
    if col in transforms:
        mu = azdias[col].mean()
        transforms[col].append(DataTransforms.Replace(None, mu))
        transforms[col].append("cast")
        transforms[col].append(min_max_scaler)
```
